```python
from Bio.PDB import *
from biopandas.pdb import PandasPdb
parser = PDBParser()
import numpy as np
from scipy.spatial import distance

#structure = parser.get_structure("3i9g_imgt", "3i9g_imgt.pdb")
```

```python
new = PandasPdb().read_pdb('3i9g_imgt.pdb')
#new1 = new.df['ATOM'].head()
kinley = new.df['ATOM']
draft = kinley[(kinley['atom_name']=='SG') &(kinley['residue_name']=='CYS') & (kinley['element_symbol']=='S')]
draft2 = draft.drop(['blank_3', 'blank_2','blank_1', 'alt_loc','insertion','blank_4','segment_id'], axis=1)
draft2
```

```python
atomic_number = draft2['atom_number'].tolist()
x_coords = draft2['x_coord'].tolist()
y_coords = draft2['y_coord'].tolist()
z_coords = draft2['z_coord'].tolist()

coords = []
for i, x, y, z in zip(atomic_number, x_coords, y_coords, z_coords):
    #print(i, x, y, z)
    coords.append([[x, y, z], i])
   # print(coords)

# Show results
for i in range(len(coords)):
    for j in range(i+1, len(coords)):
        dist = distance.euclidean(coords[i][0], coords[j][0])
        if dist <= 3:
            print('Bond length between the atoms with serial number {} and {} is {:.5f}'.format(coords[i][1], coords[j][1], dist))
```
